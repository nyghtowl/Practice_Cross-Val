Each label is scored for TPR and FPR for each label normalized by all other labels:

$TNR_{cat}$ = $\frac{ 30+5+3+22}{30+5+3+22+2+3}$ = $92.3$% Specific

$TPR_{dog}$ = $\frac{ 30}{30 + 3 + 3}$ = $83.33%$% Sensitive

$TNR_{dog}$ = $\frac{ 20+22 + 2 + 2}{20+22 + 2 + 2 + 5 + 2}$ = $86.79$% Specific

### Bias Variance and Cross validation 

We want to tune our parameters to fit our model well. 

**High Bias (poor fit):** 
    
1. We are not measuring the correct variable. 
2. We cannot correct for bias by increasing sample size
3. We measure Bias using the Accuracy metric:

$$Accuracy = \frac {TP + TN}{TP + TN + FP + FN} $$

**High variance** (Overfit model): 

1. too many parameters describe the model, 
2. the model won't be able to predict new test cases well. 
3. We can decrease Variance by increasing sample size
4. We measure Variance using the Precision metric:

$$Precision = \frac {TP}{TP + FP}$$


Cross validation Navigate the space and finds a set of Algorithms [Naive Bayes flavors in scikit learn](http://scikit-learn.org/stable/modules/naive_bayes.html) and parameters (Alpha) that can fit out model well and are robust to new data points. 

<u>Cross validation sequence:</u>

1. Split the data into 3 partitions: 

    a. Train set (60%)
    
    b. Cross validation set (20%) Tune parameters to avoid contaminating the test and train set
    
    c. Test set (20%)

2. Run the Naive Bayes algorithm on the training set. 

    a. Evaluate the Error (for Naive Bayes we use the Accuracy metric) on the cross validation set.

    b. Tune the Laplace smoothing $\alpha$ parameter to minimize the error on the cross validation set 

3. Predict labels on test set

```python
def count_accuracy(predict, actual):
    count = 0
    for i, word in enumerate(predict):
        if word == actual[i]:
            count += 1
    return count *1.0/ len(actual)
```

```python
# Naive Bayes algorithm
def plot_NB_accuracy(x_train, y_train, x_test, y_test):
    alphas = np.arange(0.001,.5,.005)
    result = []
    for a in alphas:
        mnb = MultinomialNB(alpha=a)
        mnb.fit(x_train, y_train)
        accuracy = mnb.score(x_test, y_test) # accuracy score
#         predict = mnb.predict(x_test)
#         accuracy = count_accuracy(predict, y_actual)
        result.append(accuracy)
        
    plot_accuracy = plt.plot(alphas, result, color="green")
    plt.title("Multinomial Bayes Accuracy Rate vs Alpha")
    plt.xlabel = ("alphas")
    plt.ylabel = ("Accuracy Rate")
    return mnb
```

```python
# Bernoulli Bayes algorithm 

def plot_B_accuracy(x_train, y_train, x_test, y_test):
    alphas = np.arange(0.01,.2,.01)
    result = []
    for a in alphas:
        bb = BernoulliNB(alpha=a)
        bb.fit(x_train, y_train)
        accuracy = bb.score(x_test, y_test)
#         predict = bb.predict(x_test)
#         accuracy = count_accuracy(predict, y_actual)
        result.append(accuracy)
        
    plot_accuracy = plt.plot(alphas, result, color="green")
    plt.title("Bernoulli Bayes Accuracy Rate vs Alpha")
    plt.xlabel = ("alphas")
    plt.ylabel = ("Accuracy Rate")
    return bb
```

```python
# Build each label confusion matrix
# sensativity = TP = single value at the point in matrix
# FP = row of value and other cells that are not that value
# FN = column of value and other cells not that value
# specificity = TN = all other cells not in that row or column

conf_names = ['tp', 'fp', 'fn', 'tn']

def calc_label_cm(cm, conf_names):
    len_cm = len(cm)
    results = []
    
    for i in range(len_cm):
        tp = cm[i,i]
        fp = cm[i,].sum() - tp # row
        fn = cm[:,i].sum() - tp
        tn = cm.sum() - tp - fp - fn
        results.append([tp, fp, fn, tn])
    return numpy.array(results)
    
label_cm = calc_label_cm(bb_cm, conf_names)
```

```python
# Explicitly calc rates
# TPR, - TP / TP + FN
# TNR - TN / TN + FP
# FPR - FP / TN + FP

def calc_rates(cm):
    len_cm = len(cm)
    # would be good to pass in the list of conf names nad index tp to get it
    for i in range(len_cm):
        tpr = cm[i,0] * 1.0 / (cm[i,0] + cm[i,3])
        tnr = cm[i,3] * 1.0 / (cm[i,3] + cm[i,2])
        fpr = cm[i,2] * 1.0 / (cm[i,3] + cm[i,2])
        print tpr, tnr, fpr
        
calc_rates(label_cm)
```
